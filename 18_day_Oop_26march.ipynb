```python
class bank_account():
    def __init__(self,balance):
        self.__balance = balance
    def deposite(self,amount):
        self.__balance = self.__balance + amount
    def withdraw(self,amount):
        self.__balance = self.__balance - amount
    def balance_check(self):
        return self.__balance
```

```python
b2 = bank_account(15000)
```

```python
b2.deposite(5000)
```

```python
b2.balance_check()
```

```python
b2.withdraw(2000)
```

```python
b2.balance_check()
```
