```python
def clean(i):
    r = i.split('M')
    k = i.split('k')
    try:
        
        if r[1] == '':
            
            num = r[0] 
            num = float(num) * 1000
            return num
    except Exception as e:
        pass
    try:
        if k[1] == '':
            
            k_res = float(k[0])
            return k_res
    except Exception as e:
        pass

    if i == 'Varies with device':
        return 0
    
    if type(i) == str:
        return float(i)
```

```python

clean('1600k')
```

```python
clean('7637')
```
