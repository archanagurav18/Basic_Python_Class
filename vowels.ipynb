```python
text = "Heelloooai World"
vowels = [char for char in text if char.lower() in 'aeiou']
print(vowels)  # Output: ['e', 'o', 'o']
```

```python
text = "Python is a simple programming language"
found_vowels = []
for char in text:
    if char.lower() in "aeiou":
        found_vowels.append(char)
print(found_vowels)
```
