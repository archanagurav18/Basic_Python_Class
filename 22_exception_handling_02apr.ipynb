```python
try:
    a = int(input('Write int number 1'))
    b = int(input('Write int number 2'))
except:
    print('Written wrong type of input')

try:
    result = a/b
    print(result)
except:
    print('Not possible to divide')

try:
    result = a * b
    print(result)
except:
    print('Can mutiply a with b')
```
