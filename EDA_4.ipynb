# Top 10 category installs

```python

plt.figure(figsize=(14,10))
sns.set_context('talk')
sns.set_style('darkgrid')

ax = sns.barplot(x='Installs', y='Category', data=df_cat_installs.head(10))
ax.set_xlabel('No of installs in billion')
ax.set_ylabel('Category')
ax.set_title('Top 10 category wise installs')
```

```python

plt.figure(figsize=(14,10))
sns.set_context('talk')
sns.set_style('darkgrid')

ax = sns.barplot(x='Installs', y='Category', data=df_cat_installs.head(10))
ax.set_xlabel('No of installs in billion')
ax.set_ylabel('Category')
ax.set_title('Top 10 category wise installs')
```
