/// Inheritance ///

```python
class parent():
    def singing(self):
        return 'parent is having singing skill'
```

```python
class child(parent):
    def dancing(self):
        return 'child is having dancing skill'
    
```

```python
p1 = child()
```

```python
p1.dancing()
```

```python
p1.singing()
```

```python
p2 = parent()
```

```python
p2.singing()
```

```python
class animal():
    def speak(self):
        return 'animal speak'

class dog(animal):
    def bark(self):
        return 'dog bark'
        
```

```python
d1 = dog()
```

```python
d1.speak()
```

```python
d1.bark()
```

```python
d2 = animal()
```

```python
d2.speak()
```

```python
class python():
    def programming(self):
        return 'python is used for programming'
class data_science(python):
    def data_analysis(self):
        return 'data_science is used for data analysis'
class machine_learning(data_science):
    def alogirthm(self):
        return 'machine learning is used to build algorithm'
class artificial_intelligence(machine_learning):
    def develop(self):
        return 'artificial_intelligence is used for development of new programming in ai'
class deep_learning(artificial_intelligence):
    def analys_data(self):
        return 'deep leaning is used to anylyse'
class gen_ai(deep_learning):
    def Generative_AI(self):
        return 'Generative AI is a type of artificial intelligence that creates new content'
class Agentic_ai(gen_ai):
    def Agentic(self):
        return 'Agentic AI refers to autonomous systems that can set goals, make decisions and excute'
        
```

```python
a1 = Agentic_ai()
```

```python
a1.programming()
```

```python
a1.alogirthm()
```

```python
a1.data_analysis()
```

```python
a1.develop()
```

```python
a1.data_analysis()
```

```python
a1.Generative_AI()
```

```python
a1.programming()
```

```python
a1.Agentic()
```

```python
class banking():
    def __init__(self,balance):
        self.balance = balance
class SBI(banking):
    def mutual_fund(self):
        result = self.balance/15
        return result
class ICICI(SBI):
    def stock(self,number_stock):
        if number_stock*7 <= self.balance :
             return f'Total number of shares is now: {number_stock}  and total value of shares is {number_stock * 7}'
        else :
            return 'Insufficient balance'

        
```

```python
c1 = ICICI(2500)
```

```python
c1.balance
```

```python
c1.mutual_fund()
```

```python
c1.stock(10)
```

```python
c2 = SBI(5000)
```

```python
c2.balance
```

```python
c2.mutual_fund()
```
